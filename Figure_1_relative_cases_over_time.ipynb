### Linear Graph COVID-19 Cases over time for five countries

### Figure 1 The relative cases overtime of Covid infectors (absolut Covid cases/population size)

```python
df_my_list['total_cases_ratio_in_%'] = (df_my_list['total_cases']/df_my_list['population'])*100
```

### Linear Graph : The relative cases overtime of Covid infectors (absolut Covid cases/population size) in percentage

```python
fig = px.line(df_my_list, x="date", y='total_cases_ratio_in_%', color ='location', title='COVID cases w.r.t population in percentage linear Graph')
fig.show()
```

### Logarithmic Graph : The relative cases overtime of Covid infectors (absolut Covid cases/population size) in percentage

```python
fig = px.line(df_my_list, x="date", y='total_cases_ratio_in_%', color ='location', title='COVID cases w.r.t population in percentage logarithmic Graph')
fig.update_layout(yaxis_type="log")
fig.show()
```

# Plotting the covid cases in percentage on a World Map

```python
data = dict(type='choropleth',
            colorscale = 'Viridis',
            reversescale = True,
           locations = df_my_list['location'],
            locationmode = "country names",
           z = df_my_list['total_cases_ratio_in_%'],
           text = df_my_list['location'],
           colorbar = {'title' : 'Covid Cases of countries with respect to population in percentage'},
           
           
           )
```

```python

layout = dict(title = 'COVID 19 distribution of total cases w.r.t population in %',
                geo = dict(showframe = False,projection = {'type':'mercator'})
             )
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap,validate=False)
```
